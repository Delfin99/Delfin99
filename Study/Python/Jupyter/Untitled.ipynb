```python
Intro to pyplot

    - It is a collection of commend style functions that make matplotlib work like MATLAB
```

```python
import matplotlib.pyplot as plt

plt.plot([1, 2, 3, 4])
plt.ylabel('some numbers')
plt.show()
```

```python
(Note: a single list or arrary to plot() assumes it is a sequence of y values, and automatically generates the x values
 for you. Since python ranges start with 0, the default x vector has the same length as y but starts with 0. Hence the x
 data are [0, 1, 2, 3])
```

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
plt.show()
```

```python
Formatting the style of your plot

    - There is an optional third argument which is format string that indicates the colour and line type of the plot
```

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'ro')
plt.axis([0, 6, 0, 20])
plt.show()
```

```python
axis() presents [xmin, xmax, ymin, ymax] and specifies the viewport of the axes
```

```python
import numpy as np

# evenly sampled time at 200 ms intervals
t = np.arange(0., 5., 0.2)

# red dashes, blue squares and green triangles
plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^')
plt.show()
```

```python
# numpy.arange(x, y)
# output should be 
data = {'a': np.arange(30), 'b': np.arange(30)}

plt.scatter('a', 'b', s=10, c='r' ,data=data)
plt.show()
```

```python
data = {'a': np.random.randint(0, 50, 50), 'b': np.arange(50)}
# data['c'] = np.arange(50)
data['c'] = np.random.randn(50)

plt.scatter('a', 'b', c='c', s=5, data=data)
plt.show()
```

```python
data['a'] = np.arange(30)
data['b'] = data['a'] + 10 * np.random.rand(30)
data['s'] = abs(np.random.randn(30)) * 100
# data['c'] = np.random.randint(0, 30, 30)
data['c'] = np.exp(np.random.randint(0, 30, 30))


print(data['a'])
print(data['b'])
print(data['s'])
print(data['c'])

plt.scatter('a', 'b', c='c', s='s', data=data)
plt.xlabel('entry a')
plt.ylabel('entry b')
plt.show()
```

```python
# plotting with categorical variable

names = ['A', 'B', 'C']
values = [1, 10, 100]

plt.figure(1, figsize=(9, 3))
plt.subplot(131)
plt.bar(names, values)
plt.subplot(133)
plt.scatter(names, values)
plt.subplot(132)
plt.plot(names, values)
plt.suptitle('Categorical Plotting')
plt.show()
```

```python
# Working with multiple figures and axes

def f(t): 
    return np.exp(-t) * np.cos(2*np.pi*t)

t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)

plt.figure(1)
plt.subplot(211)
plt.plot(t1, f(t1), 'bo', t2, f(t2), 'k')

plt.subplot(212)
plt.plot(t2, np.cos(2*np.pi*t2), 'r--')
plt.show()
```
